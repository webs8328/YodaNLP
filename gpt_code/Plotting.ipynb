```python
perps = [(4, [(1000, 6.80380467414856), (2000, 6.351147820949555), (3000, 5.800394930839539), (4000, 5.873440709114075)]), (8, [(1000, 6.605064406394958), (2000, 6.251836144924164), (3000, 5.838300440311432), (4000, 5.86056601524353)]), (16, [(1000, 6.563970026969909), (2000, 6.083837532997132), (3000, 5.7964745926856995), (4000, 5.66779155254364)]), (32, [(1000, 6.552771639823914), (2000, 6.238331310749054), (3000, 5.815477633476258), (4000, 5.732893242835998)]), (64, [(1000, 6.520307052135467), (2000, 6.415595407485962), (3000, 6.007957718372345), (4000, 5.867990930080413)])]

ranks = [entry[0] for entry in perps]
data = [entry[1] for entry in perps]
print(ranks)
print(data)
```

```python
steps = []
vals = []
for entry in data:
    steps.append([sub_entry[0] for sub_entry in entry])
    vals.append([sub_entry[1] for sub_entry in entry])

print(steps)
print(vals)
```

```python
for i in range(len(steps)):
    plt.plot(steps[i], vals[i], label=ranks[i])

plt.legend()
plt.title("perplexities vs steps")
plt.ylabel("perplexities")
plt.xlabel("steps")
plt.show()
```
